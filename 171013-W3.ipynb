```r
# voice.part 의 각 범주의 갯수를 세어 Freq 컬럼으로 만들어 저장

singer$Freq <- ave(singer$height, singer$voice.part, FUN = length)
```

```r
# 교차표 만들기

xtabs(formula = ~ height + voice.part, data = singer)
```

**분석**
- 아래의 그림처럼 키가 크면 저음 파트 부분이, 키가 작을수록 고음 파트가 많다.

#### 히스토그램 그리기

- 하나의 이미지에 여러개의 히스토그램을 그릴 수 있는 histogram()
- 연속형 변수 | factor(레벨이 있는 변수)
- layout = c(2행, 4열)
- nint 구간의 갯수를 몇개로 줄 것인지

```r
histogram(~ singer$height | singer$voice.part, data = singer,
         aspect = 1,
         layout = c(2, 4),
         nint = 15,
         xlab = "Height (inches)")
```

```r
qqmath(~ height | voice.part, data = singer,
      aspect = 1,
      layout = c(2, 4),
      prepanel = prepanel.qqmathline,
      panel = function(x) {
          panel.grid()
          panel.qqmathline(x)
          panel.qqmath(x)
      },
      xlab = "Unit Normal Quantile",
      ylab = "Height (inches)")
```

```r
barchart(Freq ~ height, groups = voice.part,
        data = singer,
        stack = TRUE, horizontal = FALSE,
        par.settings = list(superpose.polygon = list(col = rainbow(8))),
        auto.key = list(x = 6, y = 7, corner = c(0, 0)))
```
