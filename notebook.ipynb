```r
library(RSQLite)
library(data.table)
```

```r
sqlite <- dbDriver("SQLite")
db2016 <- dbConnect(sqlite, "dataset/lahman2016.sqlite")
```

```r
dbListTables(db2016)
```

```r
queryResult <- dbSendQuery(db2016, "SELECT * FROM Teams 
INNER JOIN TeamsFranchises
ON Teams.franchID == TeamsFranchises.franchID
WHERE Teams.G >= 150 AND TeamsFranchises.active == 'Y';")
```

```r
dat <- data.table(fetch(queryResult))
```

```r
str(dat)
```

```r
na.omit(dat)
```
